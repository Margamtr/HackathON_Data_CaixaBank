# Limpieza datos
-Quitar $ y cambiar tipo de valor a float 
-Formatear fecha. Crear columna fecha y año.
-Renombrar "id" de card_data por "card_id"

```python
df_client_data_copy=df_client_data.copy()
df_card_data_copy=df_card_data.copy()
```

```python
df_card_data_copy.rename(columns={"id":"card_id"}, inplace=True)
```

```python
df_card_data_copy.columns
```

```python
columnas_limpieza_dollar=["per_capita_income","yearly_income","total_debt","credit_limit"]
def limpieza_dollar(df,columna):
    df[columna]=df[columna].str.replace("$","")
    df[columna]=pd.to_numeric(df[columna], errors='coerce').fillna(9999).astype(np.int64)
    return df
```

```python
for col in columnas_limpieza_dollar:
    try:
        limpieza_dollar(df_client_data_copy,col)
    except:
        pass
```
